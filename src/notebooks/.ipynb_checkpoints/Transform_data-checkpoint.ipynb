```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
```

```python
def Convert_type(df):
    df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%Y")
    df = df[['Date', 'Price', 'Open', 'High', 'Low']]

    for col in ["Price", "Open", "High", "Low"]:
        df[col] = df[col].astype(str).str.replace(",", "").str.strip().astype(float)

    return df
```

```python
def plot_ts(data, label):
    sns.lineplot(data=data, x='Date', y=label)
    plt.title(label)
    plt.ylabel('')
    plt.xlabel('')
    
    min_year = data['Date'].dt.year.min()
    max_year = data['Date'].dt.year.max()
    
    tick_years = list(range(min_year, max_year + 1, 5))
    tick_dates = pd.to_datetime([f"{year}-01-01" for year in tick_years])
    
    plt.xticks(ticks=tick_dates, labels=[str(year) for year in tick_years], rotation=45)
```

```python
def Fullfill_data(df):
    df.set_index("Date", inplace=True)
    df = df.sort_index()
    df = df[~df.index.duplicated(keep="last")]

    full_date_range = pd.date_range(start=df.index.min(), end=df.index.max(), freq="D")
    df = df.reindex(full_date_range, method="ffill")

    df = df.reset_index().rename(columns={"index": "Date"})
    return df
```

```python
path = "../Data/"
folders = [f for f in os.listdir(path) if os.path.isdir(os.path.join(path, f))]
folders
```

```python
path = "../Data/"
folders = [f for f in os.listdir(path) if os.path.isdir(os.path.join(path, f))]

for folder in folders[1:]:
    folder_path = path + folder + '/'
    csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

    df_list = [pd.read_csv(file) for file in csv_files]
    df = pd.concat(df_list, ignore_index=True)
    
    df = Convert_type(df)
    df = Fullfill_data(df)
    
    print('-------------------------------------')
    print(folder)
    print(df.head())
    print(df.describe())
    
    plt.figure(figsize=(20, 15))

    for i, col in enumerate(df.columns[1:]):
        plt.subplot(4, 2, i + 1)
        plot_ts(df, col)

    plt.suptitle(folder, fontsize=14)
    plt.tight_layout()
    plt.show()
    
    output_path = path + 'Clean/' + folder.strip() + '.csv'
    print(output_path)
    
    df.to_csv(output_path)
    print('Transform completed')
    print('-------------------------------------')
    
```
